```python
with open("/content/drive/MyDrive/Hello/mine.txt", "r") as file:
    text = file.read()
```

```python
from PIL import Image
import numpy as np
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt

# Load the mask image
mask_image = np.array(Image.open("/content/drive/MyDrive/Hello/cloud.png"))

# Generate word cloud
wordcloud = WordCloud(
    width=800,
    height=400,
    background_color="white",
    mask=mask_image,
    stopwords=STOPWORDS
).generate(text)

# Display the word cloud
plt.figure(figsize=(8, 8), facecolor=None)
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```
